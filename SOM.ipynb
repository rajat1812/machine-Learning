```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Credit_Card_Applications.csv")
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc [:,-1].values
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
x = sc.fit_transform(x)
```

```python
# Training the SOM 
from minisom import MiniSom
som = MiniSom(x=10, y=10, input_len=15 ,sigma=1.0 ,learning_rate=0.5)
som.random_weights_init(x)
som.train_random(data=x , num_iteration=100)
```

```python
# Visualizing the results
from matplotlib.pyplot import bone, pcolor, colorbar, plot, show
```

```python
bone()                             # Initialize the window on which map will display
pcolor(som.distance_map().T)       # Use mid values to display colours instead of values for winning nodes
colorbar()                         # gives legend
markers = ['o','s']
colors = ['r','g']
for i,j in enumerate(x):
    w = som.winner(j)              # find out the winning node
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]], 
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2
        )

show()
```

```python
# Finding the frauds 
mappings = som.win_map(x)                                # returns dictionary of all mappings from winning nodes 
frauds = np.concatenate((mappings[(4,8)] , mappings[(3,7)]),axis= 0) 
frauds = sc.inverse_transform(frauds)
frauds
```
